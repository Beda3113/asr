```python
import os
import sys
import torch
import torchaudio
from pathlib import Path
import json
import warnings
warnings.filterwarnings("ignore")

os.chdir("/content/asr")
sys.path.insert(0, '/content/asr')

from src.model.deepspeech2 import DeepSpeech2
from src.text_encoder.ctc_text_encoder import CTCTextEncoder
from src.datasets.custom_dir_audio_dataset import CustomDirAudioDataset
from src.metrics.utils import calc_cer, calc_wer

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🚀 Устройство: {DEVICE}")

AUDIO_DIR = "/content/asr/custom_dir/audio"
TRANS_DIR = "/content/asr/custom_dir/transcriptions"

# Загрузка модели
N_TOKENS = 28
print("\n🔄 Загрузка модели...")
model = DeepSpeech2(n_tokens=N_TOKENS, n_feats=80, dim=512, gru_layers=5, dropout=0.3)
checkpoint = torch.load("checkpoint-epoch50.pth", map_location=DEVICE)
model.load_state_dict(checkpoint['state_dict'])
model.to(DEVICE)
model.eval()
print(f"✅ Модель загружена")

text_encoder = CTCTextEncoder()

# Создание датасета
print("\n🔄 Загрузка датасета...")
dataset = CustomDirAudioDataset(
    audio_dir=AUDIO_DIR,
    transcription_dir=TRANS_DIR,
    text_encoder=text_encoder,
    instance_transforms={
        "get_spectrogram": torchaudio.transforms.MelSpectrogram(
            sample_rate=16000, n_mels=80, n_fft=400, hop_length=160, power=1.0
        )
    }
)

print(f"✅ Датасет загружен. Всего примеров: {len(dataset)}")

# Инференс
results = []
print("\n🔄 Запуск инференса...\n")
with torch.no_grad():
    for i, item in enumerate(dataset):
        # spectrogram из датасета может иметь размерность:
        # - [n_feats, T] (2D) или
        # - [1, n_feats, T] (3D)
        spectrogram = item['spectrogram']

        # Если размерность 3D [1, n_feats, T], убираем лишнюю размерность
        if spectrogram.dim() == 3:
            spectrogram = spectrogram.squeeze(0)  # [n_feats, T]

        # Добавляем batch dimension: [1, n_feats, T]
        spectrogram = spectrogram.unsqueeze(0).to(DEVICE)

        spectrogram_length = torch.tensor([spectrogram.shape[-1]]).to(DEVICE)

        print(f"  DEBUG: spectrogram shape = {spectrogram.shape}")  # Должно быть [1, 80, 201]

        outputs = model(spectrogram=spectrogram, spectrogram_length=spectrogram_length)
        predicted_text = text_encoder(outputs['log_probs'], outputs['log_probs_length'])[0][0]

        results.append({
            'file': Path(item['audio_path']).stem,
            'prediction': predicted_text,
            'target': item['text']
        })

        print(f"[{i+1}/{len(dataset)}] {Path(item['audio_path']).name}: {predicted_text}")

# Сохранение результатов
SAVE_DIR = Path("/content/asr/custom_predictions")
SAVE_DIR.mkdir(exist_ok=True, parents=True)

for res in results:
    with open(SAVE_DIR / f"{res['file']}.json", 'w', encoding='utf-8') as f:
        json.dump({"prediction": res['prediction'], "target": res['target']}, f, indent=2, ensure_ascii=False)

print(f"\n✅ Сохранено {len(results)} предсказаний в {SAVE_DIR}")

# Метрики
print("\n" + "="*50)
total_cer, total_wer = 0.0, 0.0
for res in results:
    cer = calc_cer(res['target'], res['prediction'])
    wer = calc_wer(res['target'], res['prediction'])
    total_cer += cer
    total_wer += wer
    print(f"{res['file']}: CER={cer:.4f}, WER={wer:.4f}")

print("-"*50)
print(f"Средний CER: {total_cer/len(results):.4f} ({total_cer/len(results)*100:.2f}%)")
print(f"Средний WER: {total_wer/len(results):.4f} ({total_wer/len(results)*100:.2f}%)")
print("="*50)
```
